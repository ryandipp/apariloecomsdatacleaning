```python
Q1 = df_a['sale_price'].quantile(0.25)
Q3 = df_a['sale_price'].quantile(0.75)
IQR = Q3-Q1
```

```python
boxplot_min = Q1 - 1.5*IQR
boxplot_max = Q3 + 1.5*IQR

print(boxplot_max)
```

```python
df_c = df_a[(df_a['sale_price'] < boxplot_max) & (df_a['sale_price'] > boxplot_min)] 
```

```python
sns.boxplot(x = df_c['sale_price'])
```

```python
# Calculate z-scores for each numeric column
z_scores = stats.zscore(df_a['sale_price'])
print(z_scores)
```

```python
# ~code for reduce by condition
df_b=df_a[~(df_a['zscr'] > 3)]
```
